# Exercício 1 - Visualização de Dados

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv( 
'https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

```python
#Simplificando Valores
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)

#Ordenando as Barras
df = df.sort_values(by='arrecadacao', ascending=False)

#Utilizando Linhas Verticais
sns.set_style("whitegrid")

#Unificando a Cor das Barras
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='navy')

#Adicionando Valores às Barras
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

#Ajustando Títulos das Labels
plt.xlabel('Arrecadação (Milhões de Dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de Filmes')

#Removendo Bordas e Adicionando Linha de Origem
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)

#Exibindo Gráfico
plt.show()
```
